# Bertrand paradox

This is a simulation of the Bertrand paradox as defined in Lawrence C. Evans' $\textit{AN INTRODUCTION TO STOCHASTIC DIFFERENTIAL EQUATIONS}$. The paradox goes as follows: 

**Take a circle of radius 2 inches in the plane and choose a chord of this circle at random. What is the probability this chord intersects the concentric circle of radius 1 inch?**

## Solution 1

Any such chord (provided it does not hit the center) is uniquely determined by the location of its midpoint.

P(chord passes through unit circle) $=\frac{\text{area of smaller circle}}{\text{area of bigger circle}} = \frac{1}{4}$

## Solution 2

By symmetry under rotation we may assume the chord is vertical. The diameter of the large circle is 4 inches and the chord will hit the small circle if it falls within its 2-inch diameter.

P(chord passes through unit circle) $= \frac{\text{diameter of smaller circle}}{\text{diameter of bigger circle}} = \frac{1}{2}$

## Solution 3

By symmetry we may assume one end of the chord is at the far left point of the larger circle.

The angle Î¸ the chord makes with the horizontal lies between $\pm \frac{\pi}{2}$ and the chord hits the inner circle if $\theta$ lies between $\pm \frac{\pi}{3}$. 

P(chord passes through unit circle) = $\frac{\frac{2\pi}{6}}{\frac{2\pi}{2}} = \frac{1}{3}$

# Coding Solution 1

## Sampling Points in a Circle

```python
def sample_random_point(num_samples):
    length = [2*(np.random.uniform(0, 1))**.5 for i in range(num_samples)]
    angle = [np.pi * np.random.uniform(0, 2) for i in range(num_samples)]
    x1 = length * np.cos(angle)
    y1 = length * np.sin(angle)
    return Table().with_columns("x", x1, "y", y1)
```

```python
def solution_1_plot(points):
    # make a simple unit circle 
    theta = np.linspace(0, 2*np.pi, 1000)
    bigx, bigy = 2 * np.cos(theta), 2 * np.sin(theta)
    smallx, smally = np.cos(theta), np.sin(theta)
    plt.plot(bigx, bigy, linestyle='-', linewidth=2, label='Big Circle')
    plt.plot(smallx, smally, linestyle='-', linewidth=2, label='Small Circle')
    plt.plot(points.column("x"), points.column("y"), marker='o', linewidth=0, label='Samples')
    plt.ylim([-2.5,2.5])
    plt.xlim([-2.5,2.5])
    plt.grid()
    plt.show(block=True)
```

```python
random_point_example = sample_random_point(1000)
random_point_example.show(10)
```

```python
solution_1_plot(random_point_example)
```

## Approximating the probability

```python
def inside_unit_circle(row):
    return (row.item(0)**2 + row.item(1)**2) <= 1
```

```python
sample_size = 100000
solution1_points = sample_random_point(sample_size)
in_unit_array = solution1_points.apply(inside_unit_circle)
solution1_points_with_indicator = solution1_points.with_column("In unit circle", in_unit_array)
solution1_points_with_indicator.show(5)
```

```python
number_of_points_inside_unit_circle = sum(in_unit_array)
number_of_points_inside_unit_circle
```

```python
number_of_points_inside_unit_circle / sample_size
```

Approximately $\frac{1}{4}$!

# Coding solution 2

```python
def sample_random_vertical(num_samples):
    samples = make_array()
    for i in range(num_samples):
        samples = np.append(samples, np.random.uniform(-2, 2))
    return samples
```

```python
def vertical_plot(points):
    # make a simple unit circle 
    theta = np.linspace(0, 2*np.pi, 1000)
    bigx, bigy = 2 * np.cos(theta), 2 * np.sin(theta)
    smallx, smally = np.cos(theta), np.sin(theta)
    plt.plot(bigx, bigy, linestyle='-', linewidth=2, label='Big Circle')
    plt.plot(smallx, smally, linestyle='-', linewidth=2, label='Small Circle')
    [plt.axvline(x=point) for point in points]
    plt.ylim([-2.5,2.5])
    plt.xlim([-2.5,2.5])
    plt.grid()
    plt.show(block=True)
```

```python
vertical_plot(sample_random_vertical(100))
```

## Approximating the probability 

```python
sample_size = 100000
vertical_cords = sample_random_vertical(sample_size)
vertical_cords
```

```python
vertical_crossing_unit = sum(abs(vertical_cords) <= 1)
vertical_crossing_unit
```

```python
vertical_crossing_prob = vertical_crossing_unit / sample_size
vertical_crossing_prob
```

Approximately $\frac{1}{2}$!

# Coding Solution 3

```python
def sample_random_thetas(num_samples):
    samples = make_array()
    for i in range(num_samples):
        samples = np.append(samples, np.random.uniform(-np.pi/2, np.pi/2))
    return samples
```

```python
def angle_plot(thetas):
    theta = np.linspace(0, 2*np.pi, 1000)
    bigx, bigy = 2 * np.cos(theta), 2 * np.sin(theta)
    smallx, smally = np.cos(theta), np.sin(theta)
    plt.plot(bigx, bigy, linestyle='-', linewidth=2, label='Big Circle')
    plt.plot(smallx, smally, linestyle='-', linewidth=2, label='Small Circle')
    [plt.plot([-2, 2.5], [0, np.sqrt(2)*np.tan(theta) * 2.5 + 2*np.sqrt(2)*np.tan(theta)], 'k-') for theta in thetas]
    plt.ylim([-2.5,2.5])
    plt.xlim([-2.5,2.5])
    plt.grid()
    plt.show(block=True)
```

```python
angle_plot(sample_random_thetas(100))
```

```python
sample_size = 100000
angled_cords = sample_random_thetas(sample_size)
angled_cords
```

```python
angle_crossing_unit = sum(abs(angled_cords) <= np.pi/6)
angle_crossing_unit
```

```python
angle_crossing_unit/sample_size
```

About $\frac{1}{3}$!
